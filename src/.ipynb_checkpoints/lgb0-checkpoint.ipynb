```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
import lightgbm as lgb

from scipy.stats import rankdata
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import StratifiedKFold
```

```python
data = pd.read_csv('..\\data\\train.csv')
test = pd.read_csv('..\\data\\test.csv')
sub = pd.read_csv('..\\data\\sample_submission.csv')

data['Preferred_Theme'] = data['Preferred_Theme'].map({'New_UI' : 1, 'Old_UI' : 0})
```

```python
data.shape, test.shape
```

```python
State_City = {
        'California' : 2,
        'British Columbia' : 1,
        'New South Wales' : 100,
        'Western Australia' : 5,
        'West Bengal' : 3,
        'Catalonia' : 12,
        'Bavaria' : 9,
        'Tamil Nadu' : 4,
        'New Delhi' : 16,
        'Central Hungary' : 7,
        'Maharashtra' : 0,
        'Singapore' : 19,
        'Tokyo' : 13,
        'Vienna' : 11,
        'Ile-De-France' : 18,
        'Tuscany' : 20,
        'England' : 15,
        'New York' : 17,
        'Berlin' : 9,
        'Ontario' : 6,
        'Taiwan' : 8
             } 
```

```python
City = {c : i for i, c in enumerate(data['City'].value_counts().index) }
State = {s : i for i, s in enumerate(data['State'].value_counts().index) }

Gender = {'Male' : 1, 'Not_Specified' : 0, 'Female' : -1}
```

```python
data['State_City'] = data['State'].map(State_City)
test['State_City'] = test['State'].map(State_City)

data['Gender'] = data['Gender'].map(Gender)
test['Gender'] = test['Gender'].map(Gender)

data['City'] = data['City'].map(City)
test['City'] = test['City'].map(City)

data['State'] = data['State'].map(State)
test['State'] = test['State'].map(State)

data['State_City'] = np.clip(data['City'] - data['State_City'], 0, 1)
test['State_City'] = np.clip(test['City'] - test['State_City'], 0, 1)
```

```python
cat_cols = ['City', 'State', 'State_City']
num_cols = ['Age', 'Gender', 'No_of_orders_placed'] + data.columns[8:-2].to_list()
```

```python
# TIME 

data['Sign_up_date'] = pd.to_datetime(data['Sign_up_date'], errors ='coerce')
data['Last_order_placed_date'] = pd.to_datetime(data['Last_order_placed_date'], errors ='coerce')

test['Sign_up_date'] = pd.to_datetime(test['Sign_up_date'], errors ='coerce')
test['Last_order_placed_date'] = pd.to_datetime(test['Last_order_placed_date'], errors ='coerce')
```

```python
data['time'] = (data['Last_order_placed_date'] - data['Sign_up_date']).dt.days
test['time'] = (test['Last_order_placed_date'] - test['Sign_up_date']).dt.days
```

```python
data['signup'] = 100 * data['Sign_up_date'].dt.year + data['Sign_up_date'].dt.month
test['signup'] = 100 * test['Sign_up_date'].dt.year + test['Sign_up_date'].dt.month

data['Last_order'] = 100 * data['Last_order_placed_date'].dt.year + data['Last_order_placed_date'].dt.month
test['Last_order'] = 100 * test['Last_order_placed_date'].dt.year + test['Last_order_placed_date'].dt.month
```

```python
num_cols += ['time', 'signup', 'Last_order']
cols = cat_cols + num_cols
```

```python
N_folds = 20
seeds = [265, 15628, 962]

data_preds = np.zeros((data.shape[0]))
test_preds = np.zeros((test.shape[0]))

feature_importance = pd.DataFrame()

for seed in seeds:
    print('---------------------------------------')
    print('SEED : ', seed)
    print('---------------------------------------')
    
    
    data['fold'] = -1
    skf = StratifiedKFold(n_splits=N_folds, shuffle=True, random_state=seed)
    for f, (_, idxs) in enumerate(skf.split(data, data['Preferred_Theme'])):
        data.loc[idxs, 'fold'] = f
        
    for F in range(N_folds):
        
        print('FOLD : ', F)
        
        train = data[data['fold'] != F].reset_index(drop=True)
        valid = data[data['fold'] == F].reset_index(drop=True)
        
        model = lgb.LGBMClassifier(objective =  'binary', 
                                   metric= 'auc',
                                   boosting_type= 'gbdt',
                                   lambda_l1= 0.0005,
                                   lambda_l2= 9.5,
                                   num_leaves= 24,
                                   feature_fraction= 1.0,
                                   bagging_fraction= 0.95,
                                   bagging_freq= 7,
                                   min_child_samples= 50, n_estimators = 300)
        
        model.fit(train[cols], train['Preferred_Theme'],
                  early_stopping_rounds = 30,
                  eval_metric="auc",
                  eval_set=[(valid[cols], valid['Preferred_Theme'],)],
                  verbose = 50)
        
        fold_importance = pd.DataFrame()
        fold_importance["feature"] = cols
        fold_importance["importance"] = model.feature_importances_
        fold_importance["fold"] = F + 1
        feature_importance = pd.concat([feature_importance, fold_importance], axis=0)
        
        
        preds = model.predict_proba(valid[cols], num_iteration = model.best_iteration_)[:,1]
        preds = rankdata(preds)
        preds = preds / np.max(preds)
        data_preds[data[data['fold'] == F].index] += preds
        
        preds = model.predict_proba(test[cols], num_iteration = model.best_iteration_)[:,1]
        preds = rankdata(preds)
        preds = preds / np.max(preds)
        test_preds += preds
```

```python
roc_auc_score(data['Preferred_Theme'], data_preds, average = 'weighted')
```

```python
feature_importance["importance"] /= 5
cols = feature_importance[["feature", "importance"]].groupby("feature").mean().sort_values(
    by="importance", ascending=False)[:50].index

best_features = feature_importance.loc[feature_importance.feature.isin(cols)]

plt.figure(figsize=(16, 12));
sns.barplot(x="importance", y="feature", data=best_features.sort_values(by="importance", ascending=False));
plt.title('LGB Features (avg over folds)');
```

```python
sub = pd.DataFrame()
sub['CustomerID'] = test['CustomerID']
sub['Preferred_Theme'] = test_preds
sub['Preferred_Theme'] /= sub['Preferred_Theme'].max()


sub.to_csv('..\\submission_csv\\lgb0.csv', index=False)
```
